```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

with open("F1_dum.json", "r") as json_file:
    F1_dum = json.load(json_file)
with open("F1_lasso.json", "r") as json_file:
    F1_lasso = json.load(json_file)
with open("F1_ridge.json", "r") as json_file:
    F1_ridge= json.load(json_file)
with open("F1_log.json", "r") as json_file:
    F1_log = json.load(json_file)

plt.figure(figsize=(8, 6))
plt.boxplot([F1_dum, F1_lasso, F1_ridge, F1_log],
            labels=['Régression simple', 'Lasso', 'Ridge', 'Log-régression'])
# Calculate mean values for each box plot
mean_values = [np.mean(d) for d in [F1_dum, F1_lasso, F1_ridge, F1_log]]

# Add mean values to the plot
for i, mean in enumerate(mean_values, start=1):
    plt.text(i, mean, f'{mean:.2f}', ha='center', va='bottom', color='black')
plt.xlabel('Model')
plt.ylabel('Score')
plt.title('Boxplot of score distribution according to model')
plt.show()
```
